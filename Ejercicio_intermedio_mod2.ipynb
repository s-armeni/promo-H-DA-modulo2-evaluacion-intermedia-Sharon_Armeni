```python
# Ejericio parte 1

'''
Utilizando la biblioteca BeautifulSoup en Python, extrae información de la siguiente web. Debes
extraer la tabla de senderos de España con la siguiente información:
1. Identificador.
2. Denominación.
3. Itinerario.

'''
```

```python
url = 'https://es.wikipedia.org/wiki/Sendero_de_Gran_Recorrido'

# realizo la solicitud HTTP

respuesta = requests.get(url)

# verifico si la solicitud ha tenido exito

respuesta.status_code
```

```python
# traduzco el contenido httml utilizando Beautiful Soup

soup = BeautifulSoup(respuesta.text, 'html.parser')

# encuentro la tabla de senderos de España

tabla = soup.find('table', class_='wikitable')
tabla

# extraigo los encabezados de la tabla que necesito. Selecciono la etiqueta 'th'

encabezados = tabla.find_all('th')[:3]
print(encabezados)

# extraigo el texto de los encabezados, iterando por la lista encabezados creada anteriormente

encabezados_senderos = [columna.text for columna in encabezados]
print(encabezados_senderos)

# uso el metodo find_all y extraigo todas las filas de la variable tabla

filas = tabla.find_all('tr')

# contenido segunda fila
filas[1]

# creo una lista para almacenar los resultados de las filas

filas_senderos = [fila.text.split("\n")[1:-1] for fila in filas[1:]]

print(filas_senderos)

# lista final senderos de España

senderos_espania = []

for sendero in filas_senderos:
    sendero = sendero[:3]
    senderos_espania.append(sendero)

print(senderos_espania)
```
